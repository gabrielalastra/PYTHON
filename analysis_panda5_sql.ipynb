read_sql_table é uma função útil, mas funciona apenas com SQLAlchemy, um Python SQL Toolkit e Object Relational Mapper.

```
from sqlalchemy import create_engine
```

```
engine = create_engine('sqlite:////content/drive/MyDrive/ColabNotebooks/chinook.db')
connect2 = engine.connect()
```

```
DF_chi2 = pd.read_sql_table('employees', con=connect2)
DF_chi2.head()
```

```
DF_chi2 = pd.read_sql_table('employees', con=connect2,
                            index_col='EmployeeId',
                            parse_dates=[ 'BirthDate', 'HireDate'])
DF_chi2.head()
```

```
connect2.close()
```
