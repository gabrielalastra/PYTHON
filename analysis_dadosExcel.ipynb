```
#para integrar o excel
import pandas as pd
#importar o arquivo
tabela_vendas = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Vendas.xlsx")
#exibir
display(tabela_vendas)
```

```
#olhando a tabela as colunas necessarias para o faturamento de cada loja
#as colunas ID LOJA e Valor Final

#filtrar as colunas e agrupar as lojas e somar os valores fim
tabela_faturamento = tabela_vendas[["ID Loja","Valor Final"]].groupby("ID Loja").sum("Valor Final")
#para por em ordem decrescente
tabela_faturamento = tabela_faturamento.sort_values(by="Valor Final", ascending=False)
display(tabela_faturamento)
```

```
#quantidade de produtos vendidos em cada loja
tabela_vendas = tabela_vendas[["ID Loja", "Quantidade"]].groupby("ID Loja").sum("Quantidade")
display(tabela_vendas)
```
