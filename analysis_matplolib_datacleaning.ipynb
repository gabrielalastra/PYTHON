```
plt.figure(figsize=(12,6))

plt.plot(x, x**2) #parabola
plt.plot(x, -1*(x**2)) #parabola-1

#titulo
plt.title('Parabola é qndo a var esta ^2 no eixo "y"')
```

```
plt.figure(figsize=(12,6))

#subplot
plt.subplot(1,2,1) #linha, coluna e qual estamos orientando
plt.plot(x, x**2)
plt.plot([0,0,0], [-10,0,100]) #linha
plt.legend([ 'xˆ2', 'linha vertical'])
plt.xlabel('x')
plt.ylabel('xˆ2')

#subplot 2
plt.subplot(1,2,2)
plt.plot(x, -1*(x**2))
plt.plot([-18,3,10], [-10,-10,-10]) 
plt.legend(['-xˆ2', 'linha horizontal'])
plt.xlabel('x')
plt.ylabel('-xˆ2')
```

```
# cria uma figura com um eixo 
fig,axes = plt.subplots(figsize=(12,6))
```

```
#define o que vai na fig
#linhas para xˆ2
axes.plot(x, (x**2), color='red', linewidth=3, marker='*', markersize=8, label='x^2')
#linha para -xˆ2
axes.plot(x, -1*(x**2), 'b--', label='-xˆ2')

axes.set_xlabel('x')
axes.set_ylabel('x**2')
axes.set_title('parabolas dois')

axes.legend()
fig
```

```
fig, axes = plt.subplots(figsize=(12,6))

axes.plot(x, x, linestyle='solid')
axes.plot(x, x+1, linestyle='dashed')
axes.plot(x, (x+2), linestyle='dashdot')
axes.plot(x, x+3, linestyle='dotted');

axes.set_title('linhas paralelas')
```

```
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, '-og', label="solid green")
axes.plot(x, x + 1, '--c', label="dashed cyan")
axes.plot(x, x + 2, '-.b', label="dashdot blue")
axes.plot(x, x + 3, ':r', label="dotted red")

axes.set_title("linhas paralelas")

axes.legend()
```

```
n=50
x = np.random.rand(n)
y = np.random.rand(n)
color = np.random.rand(n)
area = np.pi *(20* np.random.rand(n))**2 #tamanho da bola

print(x)
print(y)
```

```
#para representar 5 dimensoes dos dados
#(valores de x e y, pontos de correlacao, tamanho e cor)
plt.figure(figsize=(14,6))
plt.scatter(x, y, s=area, c=color, alpha=0.5, cmap='Spectral')
plt.colorbar() 
#para mais opcoes erre o cmap e veja as possibilidades
#https://matplotlib.org/users/colormaps.html
plt.show()
```

```
fig = plt.figure(figsize=(14,6))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(x,y, s=area,c=color,alpha=0.5, cmap='Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x,y, s=area,c=color,alpha=1, cmap='Pastel2')
plt.colorbar()

plt.show()
```

**Histogramas**

```
value = np.random.randn(1000)
```

```
plt.subplots(figsize=(12,6))
plt.hist(value, bins=90, alpha=1, histtype='bar', color= 'steelblue', edgecolor='green')
plt.xlim(xmin=-5,xmax=5)
plt.show()
```

Estimativa de Densidade Kernel

```
from scipy import stats
```

```
densidd = stats.kde.gaussian_kde(value)
densidd
```

```
plt.subplots(figsize=(12,6))
value2 = np.linspace(min(value)-10, max(value)+10,100)
plt.plot(value2, densidd(value2), color='#FF7F00')
#representando outra dimensao dos dados
plt.fill_between(value2,0,densidd(value2+2), alpha=0.5, color='red')
plt.xlim(xmin=-5,xmax=5) #tenta tirar pra ver todos os dados

plt.show()
```

combinando plots

```
plt.subplots(figsize=(12,6))
plt.hist(value, bins=100, alpha=0.5, density=1,
         histtype='bar', color='steelblue', edgecolor='green')
plt.plot(value2, densidd(value2+2), color='#FF7F00', linewidth=3)
plt.xlim(xmin=-5, xmax=5)
plt.show()
```
